```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import SelectKBest
```

```python
data=pd.read_csv('/content/train (1).csv')
data
```

```python
data.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
data
```

```python
train_data=data.iloc[:,1:]
target_data=data.iloc[:,:1]
x_train,x_test,y_train,y_test=train_test_split(train_data,target_data,test_size=0.2,random_state=1)
```

```python
x_train.isnull().sum()
```

```python
y_train.head()
```

```python
x_train
```

```python
trf1=ColumnTransformer([
    ('impute_age',SimpleImputer(strategy='mean'),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')
```

```python
trf2=ColumnTransformer([
    ('ohc_sex_embarked',OneHotEncoder(sparse=False,handle_unknown='ignore'),[1,6])
],remainder='passthrough')
```

```python
trf3=ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,8))
])
```

```python
trf5=DecisionTreeClassifier()
```

```python
pipline=Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf5',trf5)
])
```

```python
pipline.fit(x_train,y_train)
```

```python
y_predic=pd.DataFrame(pipline.predict(x_test))
y_predic.value_counts()
```

```python
y_test.value_counts()
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipline,x_train,y_train, cv=5,scoring='accuracy').mean()
```
