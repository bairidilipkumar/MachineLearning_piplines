```python
x_train,y_train=train_data.to_numpy()[:,:-1],train_data.to_numpy()[:,-1]
x_test,y_test=test_data.to_numpy()[:,:-1],test_data.to_numpy()[:,-1]

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler,FunctionTransformer
from copy import deepcopy

std_scaler=StandardScaler().fit(x_train[:, :2])
min_max_scaler=MinMaxScaler().fit(x_train[:, 2:])

def preprocess(x):
  a=np.copy(x)
  a[:,:2]=std_scaler.transform(x[:, :2])
  a[:,2:]=min_max_scaler.transform(x[:, 2:])
  return a

```

```python
preprocess(x_test)
```

```python
preprocess_transformer =FunctionTransformer(preprocess)
preprocess_transformer
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
```

```python
p1= Pipeline([('scaler',preprocess_transformer),
              ('Linear Regression',LinearRegression())])
p1
```

```python
from sklearn.metrics import mean_absolute_error

def fit_print(p, x_train=x_train, y_train=y_train, x_test=x_test, y_test=y_test):
  p.fit(x_train,y_train)
  train_pred=p.predict(x_train)
  test_pred=p.predict(x_test)
  print("Training error...",mean_absolute_error(y_train,train_pred))
  print("Test error...",mean_absolute_error(y_test,test_pred))
```

```python
fit_print(p1,x_train, y_train, x_test, y_test)
```

```python
from sklearn.metrics import r2_score
y_pred=p1.predict(x_test)
l=r2_score(y_test,y_pred)
print(l)
```
